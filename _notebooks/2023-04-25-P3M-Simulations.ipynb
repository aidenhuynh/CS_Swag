## What are simulations by College Board definition?

- Simulations are abstractions that mimic more complex objects or phenomena from the real world
    - Purposes include drawing inferences without the contraints of the real world
- Simulations use varying sets of values to reflect the changing state of a real phenomenon
- Often, when developing a simulation, it is necessary to remove specific details or simplify aspects
    - Simulations can often contain bias based on which details or real-world elements were included/excluded
- Simulations allow the formulation of hypotheses under consideration
- Variability and randomness of the world is considered using random number generators
- Examples: rolling dice, spinners, molecular models, analyze chemicals/reactions...

## Analyzing an Example: Air-Traffic Simulator

## Card Flip

````python
import random

cards = ["Ace", "2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King"] 
suits = ["Diamonds", "Hearts", "Spades", "Clubs"]

def draw(n):
    while n > 0:
        print(random.choice(cards) + " of " + random.choice(suits))
        n -= 1
        
draw(10)
````

## Coin Flip

````python
import random

def coinflip(n):         #def function 
    tails = 0
    heads = 0
    
    while n > 0:
        randomflip = random.randint(1, 10)
        
        if randomflip <= 4: #  40% likely to choose heads
            print("Heads")
            heads += 1
            
        else:
            print("Tails")
            tails += 1
            
        n -= 1
        
    print("\nHeads: " + str(heads) + "\nTails: " + str(tails))

#Tossing the coin n times:
coinflip(10)
````

## Spin the Wheel

````python
import random

from PIL import Image


print("Spin the wheel!")
print("----------------------------------")

def wheel(n):
    blueImg = Image.open(r"images/blueWheel.png")
    redImg = Image.open(r"images/redWheel.png")
    
    blue = 0
    red = 0
        
    for i in range(n):
        
        spin = random.randint(1,2)
        if spin == 1: # head
            blue = blue + 1
            print("Blue!")
            display(blueImg)
        else:         # tail
            red = red + 1
            print("Red!")            
            display(redImg)
 
    print()
    print('Number of blue:', blue)
    print('Number of red:', red)
    
wheel(2)
````

## Population Growth and Plots

````python
import random

totalPopulation = 50 
growthFactor = 1.00005
dayCount = 0 #Every 2 months the population is reported

while totalPopulation < 1000000:
    totalPopulation *= growthFactor
    #Every 56th day, population is reported
    dayCount += 1
    if dayCount == 56: 
        dayCount = 0
        print(totalPopulation)
````

Here we initialize the total population to be 50, then set the growth factor as 1.00005 (.005 percent change). It will print the population every 56th day until it reaches one million. It multiplies the current population by the growth factor in each iteration, and increments the day count. When the day count reaches 56, it prints the current population and resets the day count to 0.

Note! This simulation assumes that the growth factor remains constant as time progresses, which may not be a realistic assumption in real-world scenarios.

### Coding!
    - League of Legends 1v1 simulation
        * This is the only database I already had downloaded for the last assignment

````python
import pandas

#read video game stats of characters and their stats to see the best ones, sorted by winrate
dataFrame = pandas.read_csv('files/lolstats.csv').sort_values(by=['Win %'], ascending=False)

def fight(): 
    # input one characters
    champ1 = input("Name a champion from league of legends (yuck!)")

    # if escape was pressed, exit function
    if champ1 == "":
        return
    
    # capitalize first letter
    champ1 = champ1.lower()
    champ1 = champ1[0].upper() + champ1[1:]
    
    # repeat the same for another character
    champ2 = input("again")
    
    if champ2 == "":
        return
    
    champ2 = champ2.lower()
    champ2 = champ2[0].upper() + champ2[1:]
    
    # try to get characters by input
    try:
        # Get the win rates of the two champions as floats without the % sign
        win1 = float(dataFrame[dataFrame.Name == champ1]["Win %"].to_string(index=False).partition('\n')[0][:-1])
        win2 = float(dataFrame[dataFrame.Name == champ2]["Win %"].to_string(index=False).partition('\n')[0][:-1])
    
    # if it doesn't work, repeat function and try again
    except:
        print("Please enter a valid character from league of legend video game")
        fight()
    
    # Show character names and their winrates
    print(champ1 + " VS " + champ2 + "\n\n" + champ1 + ": " + str(win1) + "%\n" + champ2 + ": " + str(win2) + "%\n")
    
    # If the first character has a higher win rate, they win, and vice versa
    if win1 > win2:
        print(champ1 + " WINS!")
    
    elif win2 > win1:
        print(champ2 + " WINS!")
        
    # Only if the winrates are the exact same it will result in a tie, which won't happen unless it's the same character
    else:
        print("IT'S A DRAW!")
        
fight()
````

This simulation replicates if two players were to use certain characters to fight, to decide who is more likely to win. This could be useful for betting or to choose a better character to use against your opponent.

This simulation is extremely biased because it does not account for several factors of the actual game. This simply uses the average win rate of a character to assume that it is better than the other character, which is already flawed reasoning, but on top of that, it disregards differences in the skill level of the people playing the characters.
